# barplot

- The style options in barplot are taken from the plot.
- The new features are names, names.arg, border.
- table, vector, and matrix can be used for dataset to draw barplot. (dataframe cannot be worked.)
- Recommendation: We can use dataframe to clean up data if it is already in a type of dataframe, and then convert it into applicable one (i.e. table). In this case, there should not be colnames in matrix.


```r
baseball   <- c(110, 300, 150, 280, 310, 210)
```

```r
barplot(baseball)
```

```r
barplot(baseball,
        xlab = "Sales",
        ylab = "Performance",
        names.arg = c("Jan","Feb","Mar","Apr","May","Jun"), # names = c("Jan","Feb","Mar","Apr","May","Jun"),
        main = "Baseball Sales",
        col ="red",
        ylim = c(0,350), 
        border = "green",
        density = c(10, 50, 10, 50, 50, 10)
        )
```

```r
barplot(baseball,
        xlab = "Sales",
        ylab = "Performance",
        names.arg = c("Jan","Feb","Mar","Apr","May","Jun"),
        main = "Baseball Sales",
        col ="red",
        ylim = c(0,350), 
        border = ifelse(baseball > 250, "red", "green"),
        density = ifelse(baseball > 250, 60, 20)
        )
```

# stacked or juxtaposed 

```r
baseball   <- c(110, 300, 150, 280, 310, 210)
basketball <- c(180, 200, 210, 190, 170, 100)
tennisball <- c(210, 150, 160, 210,  70, 120)
xlabel <- c("Jan","Feb","Mar","Apr","May","Jun")
ball_sales <- rbind(baseball, basketball, tennisball)
```

## stacked

```r
barplot(ball_sales)
```

## juxtaposed

```r
barplot(ball_sales,
       beside = T, # applicable to matrix
        names = xlabel
       )
```

# horizontal

```r
barplot(ball_sales,
       beside = T, # applicable to matrix
        names = xlabel,
        horiz = T
       )
```
