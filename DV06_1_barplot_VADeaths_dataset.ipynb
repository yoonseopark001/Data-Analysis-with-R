# Data Visualization with VADeaths - barplot

```r
data(VADeaths)
VADeaths
class(VADeaths)
```

```r
barplot(VADeaths)
```

```r
barplot(VADeaths,
        beside = F,
        col = rainbow(nrow(VADeaths)),
        ylim = c(0, 200),
        main = "Death Rates per 1,000 in Virginia (1940)",
        legend = T, # legend.text = T
        ylab = "Death Rates per 1,000"
       )
```

- Death rate of male is higher in Urban, while the death rates of female are similar in rural and urban

```r
barplot(VADeaths,
        beside = T,
        col = rainbow(nrow(VADeaths)),
        ylim = c(0, 80),
        main = "Death Rates per 1,000 in Virginia (1940)",
        legend = T, # legend.text = T
        ylab = "Death Rates per 1,000"
       )
```

```r
# Alternatively,

barplot(VADeaths, 
        beside = T,
        col = rainbow(nrow(VADeaths)),
        ylim = c(0, 80),
        main = "Death Rates per 1,000 in Virginia (1940)",
        ylab = "Death Rates per 1,000"
)

legend(21, 71, 
       c("50-54","55-59","60-64","65-69","70-74"), 
       cex=0.8, 
       fill = rainbow(nrow(VADeaths)))
```

```r
barplot(VADeaths,
        beside = T,
        col = rainbow(nrow(VADeaths)),
        xlim = c(0, 80),
        main = "Death Rates per 1,000 in Virginia (1940)",
        legend = T, # legend.text = T
        xlab = "Death Rates per 1,000",
        horiz = T
       )
```
