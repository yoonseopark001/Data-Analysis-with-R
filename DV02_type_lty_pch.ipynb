```r
TSdata <- read.csv("./Data/MyData.csv")
summary(TSdata)
```

```r
samp <- TSdata[1:8,]
samp
```

# Plot Style details

## type

```r
par(mfrow = c(2,2))

plot(samp$SmallGrowth ~ samp$date, # scatter plot
     type = "p",
     main = "type p")
plot(samp$SmallGrowth ~ samp$date,
     type = "l",
     main = "type l")
plot(samp$SmallGrowth ~ samp$date,
     type = "b",
     main = "type b")
plot(samp$SmallGrowth ~ samp$date,
     type = "c",
     main = "type c")
plot(samp$SmallGrowth ~ samp$date,
     type = "o",
     main = "type o")
plot(samp$SmallGrowth ~ samp$date,
     type = "h",
     main = "type h")
plot(samp$SmallGrowth ~ samp$date,
     type = "s", # left
     main = "type s")
plot(samp$SmallGrowth ~ samp$date,
     type = "S", # right
     main = "type S")
plot(samp$SmallGrowth ~ samp$date,
     type = "n",
     main = "type n")
```

## lty

```r
par(mfrow = c(2,2))

plot(samp$SmallGrowth ~ samp$date,
     type = "l",
     lty = 1, # lty = "solid"
     main = "lty = 1")
plot(samp$SmallGrowth ~ samp$date,
     type = "l",
     lty = 2, # lty = "dashed"
     main = "lty = 2")
plot(samp$SmallGrowth ~ samp$date,
     type = "l",
     lty = 3, # lty = "dotted"
     main = "lty = 3")
plot(samp$SmallGrowth ~ samp$date,
     type = "l",
     lty = 4, # lty = "dotdash"
     main = "lty = 4")
plot(samp$SmallGrowth ~ samp$date,
     type = "l",
     lty = 5, # lty = "longdash"
     main = "lty = 5")
plot(samp$SmallGrowth ~ samp$date,
     type = "l",
     lty = 6, # lty = "twodash"
     main = "lty = 6")
plot(samp$SmallGrowth ~ samp$date,
     type = "o",
     lty = 1,
     main = "lty = 1 & type o")
plot(samp$SmallGrowth ~ samp$date,
     type = "l",
     lty = 0, # lty = "blank"
     main = "lty = 0")
```
