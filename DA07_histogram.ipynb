# Histogram

```r
data(iris)
summary(iris$Sepal.Length)
```

- min: 4.3 <   > max: 7.9

```r
table(iris$Sepal.Length)
```

```r
hist(iris$Sepal.Length,
     xlim = c(4, 8),
     xlab = "Sepal Length",
     col = "aquamarine",
     main = "Histogram of Iris Sepal Length"
    )
```

```r
summary(iris$Sepal.Width)
```

```r
hist(iris$Sepal.Width,         
     xlab="Sepal Width", 
     col = "mistyrose",
     main = "iris 꽃받침 넓이 histogram", 
     xlim = c(2.0, 4.5),
     ylim = c(0,40)
    )
```
