```r
data(iris)
```

```r
plot(iris$Sepal.Length, iris$Petal.Length,
     rect(4.2,0.9, 5.9,2.1, density = 20, col = "red")
     )
text(5.1, 2.3, "This group is Setosa")

mtext("1 bottom",  side=1) # clockwise from bottom
mtext("2 left",    side=2)
mtext("3 top",     side=3)
mtext("4 right",   side=4)
```

```r
plot(iris$Sepal.Length, iris$Petal.Length)

# add rectangle in plot
rect(4.2,0.9, 5.9,2.1, density = 10, col = "blue")

# add text in plot
text(5.9, 2.2, "This group is Setosa", srt = -45, col = "dark blue")

# add text in axis
mtext("1 bottom",  side=1) # clockwise from bottom
mtext("2 left",    side=2)
mtext("3 top",     side=3)
mtext("4 right, adj = 0.08",   side=4, adj = 0.08, col = "blue")
```
