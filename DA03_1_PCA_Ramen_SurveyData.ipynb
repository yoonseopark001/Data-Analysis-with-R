```r
PCA_ramen <- prcomp(data, scale = T)
PCA_ramen
biplot(PCA_ramen)
```

    : PC1 Noddle (0.5246079) is the most influential variable in the overall evaluation of Ramen

    - RiceSariGom and Chapagetti scores well in noodle.
    - ShinRamen's package is preferred the most
    - ShinRamen, SpicyChickenFlavor, and SquidJjamppong get good scores in soup. Since Chapaghetti is not a typical ramen as it has no soup, it does not score well in soup.
    - Low scores of AnseongTangmen in all aspects are likely because it takes low cost strategy, which causes low marks in product quality survey.
