# pie

```r
weeeklySales <- c(210, 110, 400, 550, 700, 190, 150)
```

```r
pie(weeeklySales,
    init.angle=0, 
    col = rainbow(length(weeeklySales)),
    labels = c("mon","tue","wed","thr","fri","sat", "sun"),
#    clockwise = T,
    )
```

# pie3D

```r
library(plotrix)
```

```r
week <- c("mon","tue","wed","thr","fri","sat", "sun")
ratio <- round(weeeklySales / sum(weeeklySales) * 100, 1)
label <- paste(week, "\n", ratio, "%") # e.g. mon 9.7%
```

```r
pie3D(weeeklySales,
      main = "Sales Performance",
      col = rainbow(length(weeeklySales)),
      labels = label,
      explode = 0.05,
      labelcex = 1.5,
      radius = 0.9,
      theta = 0.7,
      start = 1,
     )
```
