```r
# cf.
x <- c(1,2,3,4,2,4)
y <- rep(2,6)
x; y;

table(x,y)
xy.df <- as.data.frame(table(x,y))
xy.df
plot(x,y,
    cex = 2 * xy.df$Freq, # 중첩되면 2배로 커지게 했음
    pch = 15,
    col="blue")
sunflowerplot(x,y)
```
