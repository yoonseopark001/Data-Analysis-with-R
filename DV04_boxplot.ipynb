## Sales figures

```r
baseball <- c(110, 300, 150, 280, 310)

basketball <- c(180, 200, 210, 190, 170)

tennisball <- c(210, 150, 160, 210, 70)
```

```r
boxplot(baseball, basketball, tennisball)
```

1. wide range: Sales are not stable and varies depening on season.
2. narrow range: Sales are consistent regardless of season
3. Mostly sales are going well, but there are very poor cases in some cases.

```r
boxplot(baseball, basketball, tennisball,
        horizontal = T,
        col = c("yellow","cyan","green")
       )
```
