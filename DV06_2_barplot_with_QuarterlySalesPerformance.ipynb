# Data

```r
sales_data <- c(305, 450, 320, 460, 330, 480, 380, 520)

names(sales_data) <- c("2017 1Q", "2018 1Q",      # index
                       "2017 2Q", "2018 2Q",
                       "2017 3Q", "2018 3Q", 
                       "2017 4Q", "2018 4Q")
```

# barplot

```r
barplot(sales_data,
       main = "2017 vs. 2018 Quarterly Sales Performances",
       ylim = c(0,600),
       col = c("red", "blue"), # col = rep(c("red", "blue"), 4),
       xlab = "Period",
       ylab = "Sales Performance (Unit: K USD)",
       space = (rep(c(2, 0.5), 4)),
       cex.names = 0.6 
       )       
```

```r
barplot(sales_data,
       main = "2017 vs. 2018 Quarterly Sales Performances",
       xlim = c(0,600),          # x <> y
       col = c("red", "blue"), 
       ylab = "Period",          # x <> y
       xlab = "Sales Performance (Unit: K USD)",  # x <> y
       space = (rep(c(2, 0.5), 4)),
       cex.names = 0.5, 
       horiz = T                 # convert x <> y
       )       
```
